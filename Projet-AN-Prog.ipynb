### Going through the Dataframe and replacing the empty cases with NaN, the '-' cases with False and the "Job Dissatisfaction" with True. Then, Adding a new column named "dissatisfied" to each of the Dataframes in which we're gonna put the results. Finally, we assign the results of each Dataframe to new copies

```python
def update_vals(i):
    if pd.isnull(i):
        return np.nan
    elif i == '-':
        return False
    else:
        return True
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].applymap(update_vals).any(1, skipna=False)
tafe_resignations_up = tafe_resignations.copy()
```

```python
dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction',
       'dissatisfaction_with_the_department', 'physical_work_environment',
       'lack_of_recognition', 'lack_of_job_security', 'work_location',
       'employment_conditions', 'work_life_balance',
       'workload']].any(1, skipna=False)
dete_resignations_up = dete_resignations.copy()
```

# Combine the data

#### Adding institute column to dete_resignations_up and we assign to each row the value "DETE".
#### Adding institute column to tafe_resignations_up and we assign to each row the value "TAFE"

```python
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
```

#### Combining the dataframes and we assign the result to combined.

```python
combined = pd.concat([dete_resignations_up, tafe_resignations_up], ignore_index=True)
```

#### drop any columns with less than 500 non null values

```python
combined_updated = combined.dropna(thresh = 500, axis =1).copy()
```

# Clean the Service Column

### extract the years of service from each value in the institute_service column

```python
combined_updated['institute_service_up'] = combined_updated['institute_service'].astype('str').str.extract(r'(\d+)')
```

```python
combined_updated['institute_service_up'] = combined_updated['institute_service_up'].astype('float')
```

```python
combined_updated['institute_service_up'].value_counts()
```

### Create a function that maps each year value to one of the career stages above and apply the function to the institute_service column. The result is affected to the new column "sevice_cat "

```python
def transform_service(val):
    if val >= 11:
        return "Veteran"
    elif 7 <= val < 11:
        return "Established"
    elif 3 <= val < 7:
        return "Experienced"
    elif pd.isnull(val):
        return np.nan
    else:
        return "New"
combined_updated['service_cat'] = combined_updated['institute_service_up'].apply(transform_service)
```

```python
combined_updated['service_cat'].value_counts()
```

# Perform Initial Analysis

### replace the missing values in the dissatisfied column with the most frequent value, False. Then, we'll calculate the percentage of employees who resigned due to dissatisfaction in each service_cat group and plot the results.
