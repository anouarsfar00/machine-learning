## Visualisation en utilisant un histogramme

```python
sns.set(style='ticks',palette='deep', font_scale=1.1)
#sns.set(style='darkgrid')
#sns.set(palette='deep') --> il y a d'autres dark, bright
#sns.set(font_scale=1.1) --> le style du label 
#alpha est utilisé pour régler la transparence
```

### Visualisation d'un histogramme d'age tout en montrant des groupes selon le sexe

```python
#Pandas reset_index() est une méthode pour réinitialiser l'index d'une trame de données. 
plt.hist(df[df["gender"]=="M"]["age"].reset_index(drop=True), alpha=0.6, label="Male", color="blue")
plt.legend()
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Histogram of age that shows several groups")
plt.show()
```

```python
plt.hist(df[df["gender"]=="F"]["age"].reset_index(drop=True), alpha=0.6, label="Female", color="yellow")
plt.legend()
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Histogram of age that shows several groups")
plt.show()
```

### Visualisation sous forme de nuage de points

```python
plt.scatter(df["height(cm)"], df["weight(kg)"])
plt.show()
```

### visualisation sous forme d'histogramme de tous les données de tous les attributs

```python
df.hist(figsize=(20,20))
plt.show()
```

### Visualisation boxplot

```python
sns.boxplot(x=df['gender'], y=df['age'])
plt.show()
```

```python
sns.boxplot(x ='age', y ='height(cm)', data = df)
plt.show()
```

### Visualisation sous forme de bar

```python
sns.countplot(x ='gender', data = df)
plt.show()
```

```python
sns.stripplot(x ='height(cm)', y ='weight(kg)', data = df, jitter = True, dodge = True)
plt.legend(['height(cm)', 'weight(kg)'])
plt.show()
```

```python
#sns.set_style('darkgrid')
sns.scatterplot( data=df, x='height(cm)',  y='weight(kg)', hue='gender')
plt.show()
```

```python
sns.kdeplot (data=df, x='age',hue='gender', multiple="stack")
plt.show()
```
